```python
import pandas as pd
import glob
import os
from io import StringIO

```

```python
data_files = glob.glob("data\*")
files = glob.glob('data\*.txt')
replace_word = "0"
items_avoid = '1', '2', '3', '4', '5', '6', '7', '8', '9'
```

```python
files
```

```python
def remove_space_rename(data_files):

    for i in range(len(data_files)):
        r = data_files[i].replace(" ","_")
        if( r != data_files[i]):
            print("renamed")
            os.rename(data_files[i],r)
```

```python
def replace_labels(files):
    for i in range(len(files)):
        # Read in the file
        edit_this = str(files[i])
        with open(edit_this, 'r') as file :
          filedata = file.read()

        # Replace the target string
        try:
            
            
            data = filedata
            df = pd.DataFrame([x.split(' ') for x in data.split('\n')]) # string to pd dataframe
            df_list = df[0].tolist()
            print(filedata)
            
            
            if any(i in df_list for i in items_avoid):
                #print(filedata)
                print(str(files[i]) + " Changed")
            df[0] = replace_word
            df_to_str = df.to_string(header=False, index=False)
            filedata = df_to_str
 
            
            
            
        except:
            continue

        # Write the file out again
        with open(edit_this, 'w') as file:
          file.write(filedata)
```

```python
replace_labels(files)
```
