```python
#For renaming files in directory
import os
# Function to rename multiple files
def main():
   i = 0
   path="E:/SOLARIS/HotBot_remote_and_hand/ML/Training_Videos/video_data/"
    
   for filename in os.listdir(path):
      my_dest ="Vid_" + str(i) + ".mp4"
      my_source =path + filename
      my_dest =path + my_dest
      # rename() function will
      # rename all the files
      os.rename(my_source, my_dest)
      i += 1
# Driver Code
if __name__ == '__main__':
   # Calling main() function
    try:
        main()
    except FileExistsError:
        print("File name exist")
        
```

```python
#for converting videos to frames (adjustable)
import cv2
import numpy as np
import os


def vid_2_frame(video):
    
    cap = video
    count = 0

    try:

        if not os.path.exists('data'):

            os.makedirs('data')


    except OSError:

        print ('Unable to directory ')

    while cap.isOpened():
        ret, frame = cap.read()

        if ret:

            name = './data/frame_vid_14_' + str(count) + '.jpg'

            print('Creating...' + name)

            cv2.imwrite(name, frame)
            
            if cv2.waitKey(30) & 0xFF == ord('1'):
                print(7)
            
            count += 5 # i.e. at 30 fps, this advances one second
            cap.set(1, count)
        else:
            cap.release()
            break
directory = 'E:/SOLARIS/HotBot_remote_and_hand/ML/Training_Videos/video_data/vid_14.mp4'
vid_dir = cv2.VideoCapture(directory)            
vid_2_frame(vid_dir)

```

```python
import cv2

import numpy as np

import os

cap = cv2.VideoCapture('E:/SOLARIS/HotBot_remote_and_hand/ML/Training_Videos/video_data/vid_2.mp4')

try:

    if not os.path.exists('data'):

        os.makedirs('data')

except OSError:

    print ('Unable to directory ')

currentFrame = 0

while(True):

    ret, frame = cap.read()

    name = './data/frame_vid_2_' + str(currentFrame) + '.jpg'

    print ('Creating...' + name)

    cv2.imwrite(name, frame)

    currentFrame += 1

cap.release()

cv2.destroyAllWindows()
```
