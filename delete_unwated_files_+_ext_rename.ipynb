```python
import pandas as pd
import glob
import numpy as np
import os
```

```python
ext_name = "xml"
before_ext = "jpeg"
after_ext = "jpg"

ext_path = "data\*." + str(ext_name)
before_ext_path = "data\*." + str(before_ext)
```

```python
files = glob.glob(ext_path)
rename_ext = glob.glob(before_ext_path)
```

```python
def del_unwanted(del_list):
    loop = len(del_list)
    for i in range(loop):
        
        try:
                       
            os.remove(str(del_list[i]))
            print(str(del_list[i]))
        except:
            continue
```

```python
def ext_rename(before,after):

    data_files = before
    for i in range(len(data_files)):
        r = data_files[i].replace(str(before_ext),str(after))
        if( r != data_files[i]):
            print(str(data_files[i]) + " renamed")
            os.rename(data_files[i],r)
```

```python
del_unwanted(files)
```

```python
ext_rename(rename_ext, after_ext)
```
