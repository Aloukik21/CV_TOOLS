```python
import os
from glob import glob
import shutil
import shutil, os
from sklearn.model_selection import train_test_split
# do test train splitting

PATH_DATASET = "labeled_data/*.jpg"


# find image names
image_files = glob(PATH_DATASET)
# remove file extension
image_names = [name.replace(".jpg","") for name in image_files]
# Use scikit learn function for convenience
train, test = train_test_split(image_names, test_size=0.2, shuffle=True)
```

```python


def copy_data(data, name):
   
    files = data
    path = "dataset/" + str(name)
    try:
        os.makedirs(path)
    except:
        pass
        
    for f in files:
        g = f + ".jpg"
        print(g)
        shutil.copy(f + ".jpg", path)
        shutil.copy(f + ".xml", path)
        shutil.copy(f + ".txt", path)
```

```python
copy_data(test , "test")
copy_data(train , "train")
```
